## Merge

```python
merged=features_df.merge(right=labels_df, on='fecha', how='left',).sort_values(by='fecha',ascending=False)
merged['tipo'] = merged['tipo'].fillna('IGNORE')
merged.head(3)
```

```python
tipo_sell = merged[merged['tipo']=='SELL']
tipo_sell['periodos'].hist()
print(tipo_sell['periodos'].describe())
```

```python
# Set to 24 periodos
```

```python
merged = merged[
    (merged['periodos']>=12) & 
    (merged['periodos']<=np.inf) | 
    (merged['tipo']=='IGNORE')]
```

```python
merged['tipo'].value_counts()
```

## Downsample

```python
! pip install scikit-learn
```

```python
from sklearn.utils import resample
```

```python
# Separate the class to downsample and the one to keep
df_majority = merged[merged['tipo'] == 'IGNORE'].copy()
df_minority = merged[(merged['tipo'] == 'BUY') | (merged['tipo'] == 'SELL')].copy()


majority_downsampled = resample(
    df_majority,
    replace=False,     # without replacement
    n_samples=1000,        # number of samples desired
    random_state=42     # for reproducibility
)

# Combine downsampled and original minority class
df_balanced = pd.concat([majority_downsampled, df_minority])
df_balanced.sort_values(by='fecha', ascending=False, inplace=True)

print(df_balanced['tipo'].value_counts())
df_balanced.head()
```

## Train/TEst

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
```

```python
df_balanced = df_balanced.drop(columns=['simbolo_x', 'timeframe_x', 'fecha', 'simbolo_y', 'timeframe_y', ' pr1', 'pr2', 'dif', 'periodos', 'minimo', 'fill'])
# df_balanced.columns
df_balanced.head()
```

```python
# [k for k in df_balanced.columns]
```

```python
# Encode the labels
le = LabelEncoder()
df_balanced['label_encoded'] = le.fit_transform(df_balanced['tipo'],)  # A=0, B=1, C=2
df_balanced.head()
```

```python
# Features and target
X = df_balanced.drop(columns=['tipo', 'label_encoded'])
y = df_balanced['label_encoded']
```

```python
[k for k in X.columns]
```

```python
# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Train logistic regression
clf = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=10000)
clf.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = clf.predict(X_test)

# Decode predictions to original labels (optional)
y_test_labels = le.inverse_transform(y_test)
y_pred_labels = le.inverse_transform(y_pred)

# Evaluate performance
print("Accuracy:", accuracy_score(y_test_labels, y_pred_labels))
print("Classification Report:\n", classification_report(y_test_labels, y_pred_labels))
```
